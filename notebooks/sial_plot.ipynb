```python
import numpy as np
import matplotlib.pyplot as plt

x = [192, 188/4, 168/24, 144/48, 96/96]
ys = np.array([[2.20, 0.758, 0.691], [1.44, 0.708, 0.733], [0.491, 0.379, 0.440], [0.349, 0.350, 0.165], [0.199, 0.157, 0.163]])

plt.rc("font", size=12)
plt.rc("axes", titlesize=12)
plt.rc("xtick", labelsize=14)
plt.rc("ytick", labelsize=14)
plt.rc("legend", fontsize=15)
plt.rc("figure", titlesize=12)

fig,ax = plt.subplots(1, figsize=(10, 5))

colors = ['green', 'purple', 'red']
marks = ['s', '^', 'x']
labels  = ['Water_Cubic12', 'Water_Cubic24', 'Water_Cubic48']
for y, color, mark, label in zip(ys.transpose(), colors, marks, labels):
    ax.plot(x, y, color=color, marker=mark, fillstyle='none', linestyle='None', label=label, markersize=10, markeredgewidth=2)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=False)
ax.set_xlabel("Si/Al Ratio", fontsize=18)
ax.set_ylabel(r"Diff Coeff $\left[\mathrm{Ang}^2\right]$", fontsize=18)

# Hide the right and top spines
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)

plt.tight_layout()
plt.savefig('sial_plot.png', dpi=200, facecolor='white', transparent=False)
plt.show()
```
