```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
size = 16
center = size/2
radius = size/4
radius2 = size/2.5
grid = np.zeros((size,size))

for i in range(0, size):
    for j in range(0, size):
        if (i-center)**2 + (j-center)**2 < radius2**2 and (i-center)**2 + (j-center)**2 > radius**2:
            grid[i,j] = 1

plt.rc("font", size=12)
plt.rc("axes", titlesize=12)
plt.rc("axes", labelsize=12)
plt.rc("xtick", labelsize=12)
plt.rc("ytick", labelsize=12)
plt.rc("legend", fontsize=12)
plt.rc("figure", titlesize=12)

circle1 = plt.Circle((center, center), radius, color='b', fill=False, lw=4)
circle2 = plt.Circle((center, center), radius2, color='b', fill=False, lw=4)
fig,ax = plt.subplots(1, figsize=(5, 5))

ax.imshow(grid, cmap=plt.get_cmap('BuGn'), alpha=0.4)
ax.set_xticks(np.arange(-.5, size, 1), minor=True)
ax.set_yticks(np.arange(-.5, size, 1), minor=True)
ax.grid(which='minor', color='k', linestyle='-', linewidth=0.3)
ax.add_patch(circle1)
ax.add_patch(circle2)

plt.tight_layout()
plt.savefig('rdf_grid.png', dpi=400)
plt.show()
```
